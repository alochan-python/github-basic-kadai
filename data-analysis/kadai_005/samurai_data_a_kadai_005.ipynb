```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots(2,3,tight_layout=True)

# 折れ線グラフ
x0 = [0, 1,  2, 3]
y0 = [1, 2, -5, 2]
ax[0,0].plot(x0,y0)

# sin関数
x1=np.linspace(0,10,100)
y1=2+2*np.sin(2*x1)
ax[0,1].plot(x1,y1)

# ヒストグラム
x2=np.random.normal(15,5,1000)
ax[0,2].hist(x2,bins=10)

x3=np.random.normal(3,2,100)
y3=np.random.normal(4,2,100)
ax[1,0].scatter(x3,y3)
plt.show()
```
