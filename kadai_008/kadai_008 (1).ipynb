```python
import random

var = random.randint(1,20)
print(var)

#====and使う方法=
if var%3 == 0 and var%5 == 0:
    print("FizzBuzz")
elif var%3 == 0:
    print("Fizz")
elif var%5 == 0:
    print("Buzz")
else:
    print(var)

#====and使わない=
if var%3 == 0:
    if var%5 == 0:
        print("FizzBuzz")
    else:
        print("Fizz")
elif var%5 == 0:
    print("Buzz")
else:
    print(var)
```
