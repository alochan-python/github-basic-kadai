```python
#名前(name)と年齢(age)の属性を持つHumanクラスを作成
class Human:
  def __init__(self, name ,age):
    self.name = name
    self.age = age
#Humanクラスには、以下の条件で標準出力(print)するcheck_adultメソッドを追加。
#ageが20以上の場合に大人であること
#そうでない場合に大人でないこと
  def check_adult(self,name,age):
    if (self.age) >=20:
      print(f"{self.name}は大人。年齢は{self.age}です。")
    else:
      print(f"{self.name}は大人ではありません。年齢は{self.age}です。")

#該当リスト４件
Menbers = {"samuraiA":10,"samuraiB":19,"samuraiC":20,"samuraiD":120}

#該当リスト４件文の判定出力
for name,age in Menbers.items():
  instance_Human = Human(name,age)
  instance_Human.check_adult(name,age)
```
